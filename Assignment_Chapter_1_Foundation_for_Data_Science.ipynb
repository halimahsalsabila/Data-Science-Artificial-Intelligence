###3 - Program BMI

```python
print("Body Mass Index (BMI) Program by Kompi 5, Group 2\n")

nama = (str(input("Sebutkan nama Anda: ")))
bb = int(input("Masukan berat badan Anda: "))
tb = int(input("Masukan tinggi badan Anda: "))
exponent = 2

tb_meter = tb / 100
BMI = round(bb/ tb_meter**exponent, 2)

if BMI < 18.5:
  print(nama + ", Skor BMI Anda adalah ", BMI , ", Berat Badan Anda termasuk kategori Underweight")
elif BMI <= 24.9:
  print(nama + ", Skor BMI Anda adalah " , BMI , ", Berat Badan Anda termasuk kategori Normal")
elif BMI <= 29.9:
  print(nama + ", Skor BMI Anda adalah " , BMI , ", Berat Badan Anda termasuk kategori Overweight")
elif BMI <= 34.9:
  print(nama + ", Skor BMI Anda adalah " , BMI , ", Berat Badan Anda termasuk kategori Obesitas")
else:
  print(nama + ", Skor BMI Anda adalah " , BMI , ", Berat Badan Anda tidak normal:)")
```

###5 - menambahkan status BMI masing-masing peserta pada dataframe

```python
#menentukan batas kategori BMI
bins = [0, 18.5, 25.0, 30.0, float('inf')]
labels = ['Underweight', 'Normal', 'Overweight', 'Obesity']

# membuat kolom 'Status BMI' berdasar pada kategori BMI
data['Status BMI'] = pd.cut(data['BMI'], bins=bins, labels=labels, right=False)

# Menampilkan dataframe dengan kolom 'Status BMI' yang baru
display(data)
```

###8 - menambahkan kolomm nomor peserta

```python
# Mengatur ulang indeks dataframe dan menambahkan 1 untuk nomor urut dimulai dari 1
data['No.'] = data.reset_index(drop=True).index + 1

# Mengatur kolom 'No.' di paling kiri dataframe
data = data[['No.'] + [col for col in data.columns if col != 'No.']]

# Menampilkan dataframe yang telah ditambahkan nomor urut
display(data)
```

```python
### nomor 2

print("Perusahaan Investasi Jamin-Sugih Co kami memberikan penawaran investasi terbaik dengan pembagian keuntungan tetap sebesar 25% setiap tahun.")

def investasi():
    modal = float(input("Masukan Modal Investasi (Rp0 - Rp1.000.000.000)\t")) #--> perintah memasukan nominal
    tenure = int(input("Masukan Jangka Waktu Investasi (1-25 Tahun)\t")) #--> perintah memasukan jangka waktu investasi
    rate = 0.25 #--> keuntungan yang akan didapatkan jika berinvestasi
    investasi_awal = modal
    investasi = 0

    if modal >= 0 and modal <= 1000000000 and tenure >= 1 and tenure <= 25: #--> kondisional
        for i in range(tenure + 1):
            investasi += modal
            total_keuntungan = investasi_awal * rate #--> logika matematika total keuntungan
            investasi += total_keuntungan #--> total keuntungan yang didapat
            print(f"\nTahun = " + str(i+1) + "\t" + "Keuntungan anda adalah " +  str(round(investasi)))

            if i == tenure-1: #--> loop akan berhenti jika i = tahun yang diinput
                break
    else:
        print("modal atau tenure tidak sesuai ketentuan") #--> kondisi jika tidak memenuhi syarat

investasi()
```
